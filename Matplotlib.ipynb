```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3], [3,4,7])
plt.title("My Plot")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
x=[1,2,3]
y=[21, 33, 42]
x2=[1.5,2.5,3.5]
y2=[25,35,40]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
x=[1,2,3]
y=[21, 33, 42]
x2=[1.5,2.5,3.5]
y2=[25,35,40]
plt.plot(x,y, label="first", linewidth=8)
plt.plot(x2,y2, label="second")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
x=[1,2,3]
y=[21, 33, 42]
x2=[1.5,2.5,3.5]
y2=[25,35,40]
plt.hist(x, label="first")
plt.scatter(x2,y2, label="second")
plt.legend()
plt.show()
```
